```r
library(leaflet)
library(htmlwidgets)
library(IRdisplay)

library(leaflet)
m = leaflet() %>% addTiles()
saveWidget(m, 'demo1.html', selfcontained = TRUE)
display_html('<iframe src="demo1.html" height="400" width="600"></iframe>')
```

```r
library(leaflet)
library(webshot)
library(htmlwidgets)
m = leaflet() %>% addTiles()

saveWidget(m, 'demo2.html', selfcontained = TRUE)
webshot("demo2.html", file = "Rplot.png")
library(png)
img <- readPNG("Rplot.png")
options(repr.plot.width = 15, repr.plot.height = 13, repr.plot.res = 100)
plot(NA, xlim = c(0, 7), ylim = c(0, 7), type = "n", xaxt = "n", yaxt = "n", xlab = "", ylab = "")
rasterImage(img, 0, 0, 7, 7)
```
